```python
# remove columns that start with "d_"
for col in list(dirty_df.columns):       # wrap in list() so dropping inside loop wonâ€™t mess up the iterator
    if col.startswith("d_"):
        dirty_df.drop(col, axis=1, inplace=True)
```

```python
def decode_if_bytes(x):
    if isinstance(x, (bytes, bytearray)):
        return x.decode('utf-8')     # or whatever encoding you need
    return x

dirty_df['gender'] = dirty_df['gender'].apply(decode_if_bytes)
```

```python
dirty_df.head()
```

```python
dirty_df['gender'] = (
    dirty_df['gender']
    .astype(str)
    .str.lstrip("b'")   # removes any leading b or '
    .str.rstrip("'")    # removes trailing '
)
```

```python
for x in dirty_df.columns:
    dirty_df[x] = (
    dirty_df[x]
    .astype(str)
    .str.lstrip("b'")   # removes any leading b or '
    .str.rstrip("'")    # removes trailing '
)
```

```python
# 1. Examine % missing per column
miss_pct = dirty_df.isna().mean() * 100
print(miss_pct.sort_values(ascending=False))

# 2. Drop ultra-sparse columns (>50% missing)
cols_to_drop = miss_pct[miss_pct > 50].index
df2 = dirty_df.drop(columns=cols_to_drop)

# 3. Drop rows only if theyâ€™re still very sparse (>30% missing)
row_thresh = int(df2.shape[1] * 0.3)
df3 = df2.dropna(thresh=df2.shape[1] - row_thresh)

print("After cleanup:", df3.shape)
```
